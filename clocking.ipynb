```python
import time
import sqlite3
from IPython.display import clear_output

# Function to display the current time for a limited period
def display_time(duration=10):
    """Display the current time in the console for a given duration (in seconds)."""
    start_time = time.time()
    while time.time() - start_time < duration:
        current_time = time.strftime('%H:%M:%S')  # Get the current time
        clear_output(wait=True)  # Clear the output for real-time effect
        print(f"Current Time: {current_time}")  # Display the current time
        time.sleep(1)  # Update every second

# Function to save the current time to the database
def save_time():
    """Save the current time to the database."""
    current_time = time.strftime('%H:%M:%S')
    cursor.execute("INSERT INTO time_history (time_stamp) VALUES (?)", (current_time,))
    conn.commit()
    print(f"\nTime '{current_time}' saved to history!")

# Function to view saved time history
def view_history():
    """View the saved time history from the database."""
    cursor.execute("SELECT * FROM time_history")
    all_times = cursor.fetchall()
    if all_times:
        print("\nTime History:")
        for record in all_times:
            print(f"ID: {record[0]}, Time: {record[1]}")
    else:
        print("\nNo history found!")

# SQLite database setup
conn = sqlite3.connect("digital_clock.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS time_history (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    time_stamp TEXT NOT NULL
)
""")
conn.commit()

# Menu for console-based interaction
try:
    while True:
        print("\nOptions:")
        print("1. Display Current Time (for 10 seconds)")
        print("2. Save Current Time")
        print("3. View Saved Time History")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            display_time(duration=10)  # Display the current time for 10 seconds
        elif choice == '2':
            save_time()  # Save the current time
        elif choice == '3':
            view_history()  # View saved history
        elif choice == '4':
            print("Exiting the application...")
            break
        else:
            print("Invalid choice! Please try again.")
finally:
    conn.close()  # Close the database connection
```

```python
import time
import sqlite3
from IPython.display import clear_output

# Function to display the current time for a user-specified duration
def display_clock():
    """Display a real-time clock in the console for a user-specified duration."""
    try:
        duration = int(input("Enter the number of seconds to display the clock: "))
        start_time = time.time()
        while time.time() - start_time < duration:
            clear_output(wait=True)  # Clear the output for a real-time effect
            current_time = time.strftime('%H:%M:%S')  # Get the current time
            print(f"Current Time: {current_time}")  # Print the current time
            time.sleep(1)  # Pause for a second
        print(f"Clock stopped after {duration} seconds.")
    except KeyboardInterrupt:
        print("\nClock stopped early.")
    except ValueError:
        print("Invalid input! Please enter a valid number.")

# Function to save the current time to the database
def save_time():
    """Save the current time to the SQLite database."""
    current_time = time.strftime('%H:%M:%S')
    cursor.execute("INSERT INTO time_history (time_stamp) VALUES (?)", (current_time,))
    conn.commit()
    print(f"Time '{current_time}' saved successfully!")

# Function to view the saved time history
def view_history():
    """Fetch and display the saved time history from the database."""
    cursor.execute("SELECT * FROM time_history")
    records = cursor.fetchall()
    if records:
        print("Saved Time History:")
        for record in records:
            print(f"ID: {record[0]}, Time: {record[1]}")
    else:
        print("No history available.")

# SQLite database setup
conn = sqlite3.connect("digital_clock.db")  # Connect to the SQLite database
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS time_history (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    time_stamp TEXT NOT NULL
)
""")  # Create the table if it doesn't exist
conn.commit()

# Main Menu
while True:
    print("\nOptions:")
    print("1. Start Real-Time Clock")
    print("2. Save Current Time")
    print("3. View Saved Time History")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        print("Press Ctrl+C to stop the clock.")
        display_clock()  # Display the real-time clock for a user-specified duration
    elif choice == '2':
        save_time()  # Save the current time to the database
    elif choice == '3':
        view_history()  # Display the saved history
    elif choice == '4':
        print("Exiting the application...")
        break
    else:
        print("Invalid choice! Please try again.")

# Close the database connection
conn.close()
2
```
